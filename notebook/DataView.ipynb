```python
def build_time_f_str(x):
    str_arr = x.split(".")
    year = int(str_arr[0])
    month = 0
    if len(str_arr)>1:
        month = int(str_arr[1])
    return year*12 + month   
```

```python
build_time = build_time.str.replace("年",".")
build_time = build_time.str.replace("ヶ月","")
build_time = build_time.str.replace("新築","0")
# 築（月）数　　大きければ大きいほど悪い
build_time = build_time.map(build_time_f_str)
```

```python
def take_min_number_in_text(text):
    pattern=r'([0-9]*)'
    lists=re.findall(pattern,text)
    number_list = []
    for item in lists:
        if item.isdigit():
            number_list.append(item)
    return int(min(number_list))       
```

```python
access = access.map(take_min_number_in_text)
```

```python
text = "専用バス／\t専用トイレ／\tバス・トイレ別／\tシャワー／\t追焚機能／\t浴室乾燥機\t／\t温水洗浄便座／\t洗面台独立／\t脱衣所'"
def bath_toilet_f(text):
    arr = text.split("／\t")
    point_map = {"専用バス":1,
                 "専用トイレ":1,
                 "バス・トイレ別":1,
                 "シャワー":1,
                 "追焚機能":1,
                 "浴室乾燥機":1,
                 "温水洗浄便座":1,
                 "洗面台独立":1,
                 "脱衣所":1,
                 "共同トイレ":-1,              
    }
    total_point = 0
    for item in arr:
        if item in point_map.keys():
            total_point += point_map[item]
    return total_point
```

```python
bath_toilet = bath_toilet.map(bath_toilet_f)
```
